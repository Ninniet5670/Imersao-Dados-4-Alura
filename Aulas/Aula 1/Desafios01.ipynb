```python
import pandas as pd
url = "https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv"
dados = pd.read_csv(url)
```

```python
dados
```

# DESAFIO 1
Média das metragens dos imóveis por bairro

```python
bairro_media = dados.groupby('Bairro').mean()['Metragem']
bairro_media
```

# DESAFIO 4
Conseguir o valor mínimo, máximo, mediana e média da coluna "Metragem"

```python
outras_informacoes_metragem = dados.groupby('Metragem').Metragem.agg(['min', 'max', 'median', 'mean'])
outras_informacoes_metragem
```

# DESAFIO 5
Identificar quais bairros não tem rua (valor nulo)

```python
bairros_sem_rua = dados[pd.isnull(dados.Rua)].Bairro.value_counts()
bairros_sem_rua
```

escrevendo o `.unique()` inves do `.value_counts()` retorna a lista completa com todos os bairros sem ruas, no entanto, foi preferível o segundo método devido ao fato dele também mostrar a quantidade de ocorrências de falta de ruas nos bairros
